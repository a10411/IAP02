```python
import pandas as pd
import seaborn as sns

insurance_filepath = "insurance.csv"

insurance_data = pd.read_csv(insurance_filepath)

insurance_data.head()

```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
#  possvel entender que existe relaao entre bmi maior -> charge maior
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
# agora  possvel entender que existe uma relao muito grande entre smoker e charge superior
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
# a linha de regresso para fumadores  muito mais acentuada que a de no fumadores
```

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```
